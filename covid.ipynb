```python
import pandas as pd
data = pd.read_csv("BCG_COVID-19_clinical_trials-2020_06_06.csv")
print(data.head())
```

```python
import matplotlib.pyplot as plt
phase_country=data.groupby("Phase")["Country"].value_counts()
print(phase_country)
count=data["Phase"].value_counts()
plt.pie(count,labels=count.index,autopct="%1.1f%%")
plt.title("Top 10 Countries with BCG Clinical Trials")
plt.show()
```

```python
import pandas as pd
data=pd.read_csv("task_2-owid-covid-data-22_September_2020.csv")
data.head(10)
```

```python
death = data[data['continent'].notnull()].copy()
```

```python
location = death.groupby('continent')['total_deaths'].sum()
```

```python
print(location)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
location.plot(kind='bar')
plt.title('Total Deaths by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Deaths')
plt.show()
```
